```python
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
import random
```

```python
root_path = Path('/userhome/bigdata/train/001')
files = list(root_path.iterdir())
```

```python
for i,file in enumerate(files):
    print(file)
    if i == 10:
        break
```

```python
def vis9img(files):
    random.shuffle(files)
    plt.subplots(figsize=(10, 10)) 
    for i,img in enumerate(files,1):
        img = plt.imread(img)
        plt.subplot(3,3,i)
        plt.axis('off')
        plt.imshow(img)
        if i == 9:
            break
    plt.show()
```

```python
vis9img(files)
```
