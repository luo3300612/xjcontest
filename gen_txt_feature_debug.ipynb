```python
def get_vector(string):
    out = np.zeros((24,))
    time_list = get_time_list(string)
    for time in time_list:
        hour_list = time.split('|')
        for hour in hour_list:
            out[int(hour)] += 1
    return out

def get_time_list(string):
    day_list = string.split(',')
    time_list = [day.split('&')[-1] for day in day_list]
    return time_list
```
